# 5.22 MATPLOTLIB
Matplotlib merupakan salah satu Library pada Python yang fungsi utamanya adalah dalam hal menggambarkan grafik, histogram dan berbagai jenis plot hasil dari pengolahan data menggunakan Library lainnya yaitu Pandas dan Numpy. Untuk itu penggunaan ketiga Library tersebut biasanya menjadi satu kesatuan. Untuk itu hal pertama yang perlu dilakukan adalah memanggil ketiga jenis Library tersebut seperti

Berikut contoh latihannya:

```python
#hal pertama yang dilakukan adalah memanggil Library
import matplotlib.pyplot as plt
```

# 5.22.1 Grafik sederhana dengan Matplotlib

```python
#latihan 1 (Grafik Garis)

#pertama-tama kita tentukan dulu angka yang berada dalam variabel x dan y
x = [5,6,7,8]
y = [6,7,8,9]
#lalu kita menggunakan plt.plot(x,y) untuk meletakkan posisi titik di x dan y
plt.plot(x,y)

#selanjutnya kita bisa memberikan judul grafik tersebut menggunakan plt.title
plt.title('Grafik y = x + 1')
#kemudian kita memberikan keterangan pada sumbu x(horizontal) menggunakan plt.xlabel
plt.xlabel('X')
#dan memberikan keterangan pada sumbu y(vertikal) menggunakan plt.ylabel
plt.ylabel('Y')
```

# 5.22.2 Properti Pembuatan Grafik dengan Matplotlib

```python
#latihan 1 (Grafik Garis)

#pertama-tama kita tentukan dulu angka yang berada dalam variabel x dan y
y = [25, 30, 35, 40, 50]
x = [2010, 2011, 2012, 2013, 2014]
#lalu kita menggunakan plt.plot(x,y) untuk meletakkan posisi titik di x dan y
plt.plot(x,y)

#kemudian kita memberikan keterangan pada sumbu x(horizontal) menggunakan plt.xlabel
plt.xlabel('Tahun')
#dan memberikan keterangan pada sumbu y(vertikal) menggunakan plt.ylabel
plt.ylabel('Jumlah siswa')

#selanjutnya kita bisa memberikan judul grafik tersebut menggunakan plt.title
plt.title('Lulusan SMA Negeri 2 Mengwi')
#kemudian kita gunakan plt.grid untuk menampilkan garis grid
plt.grid(True)

#dan kita menggunakan plt.show untuk menampilkan grafik
plt.show()
```

```python
#latihan 2 (Grafik Batang)

#pertama-tama kita tentukan dulu angka yang berada dalam variabel x dan y
x = [1, 2, 3, 4, 5]
y = [80, 90, 95, 85, 79]

#lalu kita gunakan fig, ax untuk menambahkan kolom figure dan axes pada sublplots()
fig, ax = plt.subplots()

#kemudian kita gunakan ax.bar untuk membuat output persegi
ax.bar(x, y, color= 'tan')

#lalu kita gunakan ax.set_title untuk membuat judul grafik
ax.set_title('Nilai Siswa Kelas A SMAN 2')
#lalu kita gunakan ax.set_ylabel untuk menamai garis vertikal dan ax.set_xlabel untuk menamai garis horizontal
ax.set_ylabel('Nilai')
ax.set_xlabel('Mata Pelajaran')
#lalu kita gunakan ax.set_xticks untuk memberikan nilai pada garis horizontal
ax.set_xticks(x)
#kemudian ada ax.set_xticklabels untuk menamai nilai yang diciptakan xticks
ax.set_xticklabels(("IPA", "IPS", "Matematika", "Bahasa Inggris", "Kesenian"))

#lalu terakhir kita gunakan plt.show untuk menampilkan grafik
plt.show()
```

```python
#latihan 3 (Grafik/Diagram Lingkaran)

#pertama kita tentukan dulu isi label untuk setiap data
labels = ['Tidak Bekerja', 'Guru', 'Dokter', 'Pengusaha']
#kemudian kita gunakan quantity untuk menentukan kuantitas dari setiap label
quantity = [9, 5, 6, 15]
#setelah itu kita gunakan colors untuk menentukan warna nya
colors = ['grey', 'plum', 'lightcoral', 'aquamarine']

#lalu kita gunakan plt.title untuk membuat judul grafik
plt.title('Pekerjaan Orang Tua Siswa SMAN 2')
#lalu kita gunakan plt.pie merupakan fungsi untuk mengidentifikasi label, kuantitas, dan warna, shadow digunakan untuk memberikan efek bayangan, startangle=90 digunakan untuk membuat tabel pada nilai pertama senilai 90 derajat
plt.pie(quantity, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

#setelah itu plt.axis untuk memberikan aspek rasio pada setiap nilai
plt.axis('equal')

#terakhir plt.show untuk menampilkan grafik
plt.show()
```

# 5.22.4 Properti Pembuatan Histogram dengan Matplotlib

```python
#latihan 1

#pertama-tama kita tentukan dulu isi label untuk setiap data
penyanyi = ["Sarwendah", "Tiara", "Ziva"]
penggemar = [100,120,108]

plt.figure(figsize=(8,4))
#kemudian tentukan warna yang ingin digunakan
histogram = plt.bar(penyanyi,penggemar,color='grey')

#selanjutnya tentukan pola yang ingin digunakan setiap datanya
histogram[0].set_hatch('x')
histogram[1].set_hatch('/')
histogram[2].set_hatch('o')

#terakhir akan ditampilkan histogramnya
plt.show()
```

# Tambahan untuk Data Bunga Iris menggunakan Pandas dan Matplotlib

```python
#hal pertama yang kita lakukan yaitu memanggil library pandas dan matplotib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#latihan 1

#pertama-tama lakukan pemanggilan data dengan menggunakan metode .read_csv 
data_iris = pd.read_csv('bunga_iris.csv')

#kemudian kita gunakan kolom petal_length dari dataframe iris_data
petal_length = data_iris['petal length (cm)']

#selanjutnya kita membuat histogram dari data petal_length
plt.hist(petal_length, bins=150, color='black')
#Akan dibuat suatu plot histogram dari petal length
#bins adalah interval dari histogram yang akan dibuat

plt.xlabel('Petal Length (cm)')
#label sumbu x akan menunjukkan petal length 
plt.ylabel('Frekuensi')
#label sumbu y akan menunjukkan frekuensi nilai panjang kelopak dalam setiap interval (bin)
plt.title('Histogram Petal Length(cm)')
#dibuat judul dengan string "Histogram Petal Length(cm)"

#terakhir kita tampilkan histogramnya
plt.show()
```

```python
#latihan 2

#pertama-tama kita lakukan pemanggilan data dengan menggunakan metode .read_csv
iris_data = pd.read_csv('bunga_iris.csv')

#Selanjutnya kita pilih kolom yang akan digunakan dalam gambar
#dalam latihan ini, dipilih kolom 'petal length (cm)' dan kolom 'petal width (cm)'
petal_length = iris_data['petal length (cm)']
petal_width = iris_data['petal width (cm)']

#selanjutnya, kita membuat scatter plot mengenai hubungan antara petal length dan petal width
plt.scatter(petal_length, petal_width)

plt.xlabel('Petal Length (cm)')
#digunakan label sumbu x untuk merepresentasikan label yang ingin ditambahkan pada sumbu x
plt.ylabel('Petal Width (cm )')
#digunakan label sumbu y untuk merepresentasikan label yang ingin ditambahkan pada sumbu y

plt.title('Scatter Plot dari Petal Length (cm) dan Petal Width (cm)')
#gunakan plt.title untuk menambahkan judul pada sactter plot dengan mengambil string Scatter Plot dari Petal Length (cm) dan Petal Width (cm)' sebagai argumen

#gunakan plt.show untuk menampilkan sactter plot
plt.show
```
